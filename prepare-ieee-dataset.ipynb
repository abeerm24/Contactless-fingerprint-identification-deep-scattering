```python
import os
```

```python
# Path to the IEEE dataset (for accessing training and gallery images)
PATH = "D:/Cross_Fingerprint_Images_Database/processed_contactless_2d_fingerprint_images/first_session"

gallery_labels = os.listdir(PATH)

train_labels = []
train_paths = []
gallery_paths = []

for label in gallery_labels:
    folder = os.path.join(PATH,label)
    files = os.listdir(folder)
    gallery_paths.append(os.path.join(folder,files[0])) # Use the first image for each fingerprint as the gallery image
    
    train_labels.extend([label for _ in range(len(files) - 1)])
    train_paths.extend([os.path.join(folder,files[i]) for i in range(1,len(files))])
```

```python
# Creating test dataset
PATH2 = "D:/Cross_Fingerprint_Images_Database/processed_contactless_2d_fingerprint_images/second_session"

test_labels = []
test_paths = []

for label in os.listdir(PATH2):
    folder = os.path.join(PATH2,label)
    files = os.listdir(folder)
    test_labels.extend([label for _ in range(len(files))])
    test_paths.extend([os.path.join(folder,files[i]) for i in range(len(files))])
```
